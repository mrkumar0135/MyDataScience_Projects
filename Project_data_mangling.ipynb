```python
users =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv' ) # reading users from csv file
sessions =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv' ) # reading session from csv file
products =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv' ) # reading products from csv file
transactions =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv') # reading transactions from csv file
temp_df=pd.merge(transactions,users,how='left',on='UserID') # applying left merge on df transactions and users and store in temp_df
temp_df # printing temp_df
```

```python
temp_df=transactions[~transactions['UserID'].isin(users['UserID'])] #selectiong rows for whom transactions.UserID not in users.UserID
temp_df #printing the temp_df
```

```python
temp_df=pd.merge(transactions,users,how='inner',on='UserID') # applying inner merge on df transactions and users and store in temp_df
temp_df # printing temp_df
```

```python
temp_df=pd.merge(transactions,users,how='outer',on='UserID') # applying inner merge on df transactions and users and store in temp_df
temp_df # printing temp_df
```

```python
pd.merge(left=users, right=sessions, how='inner', 
         left_on = ['UserID', 'Registered'], 
         right_on = ['UserID', 'SessionDate']) # merging df users and session for common userID and Registered == SessionDate
```

```python
user_id_df=pd.DataFrame({'UserID':users['UserID'].values,'temp':np.ones(len(users)).astype(int)})  #making a new df containing users.UserID
products_id_df=pd.DataFrame({'ProductID':products['ProductID'].values,'temp':np.ones(len(products)).astype(int)}) #make a new df containing products.ProductID
cart_pr_df=pd.merge(user_id_df,products_id_df,how='inner')[['UserID','ProductID']] # merging and performing slicing at the same time
cart_pr_df
```

```python
output_df=pd.merge(left=cart_pr_df,right=transactions,how='left',
                   left_on=['UserID','ProductID'],
                   right_on=['UserID','ProductID']) # merging cart_pr_df and transactions on left outer
output_df=output_df.fillna(0) # fill nan with 0
cart_pr_df['Quantity']=output_df.groupby(['UserID','ProductID']).Quantity.sum().values # adding a column Quantity and storing value
cart_pr_df # printing the data frmame
```

```python
data=pd.merge(users, transactions.groupby('UserID').first().reset_index(),how='left', on='UserID')
my_columns = list(data.columns)
my_columns
```

```python
list(data.dropna(thresh=int(data.shape[0] * .9), axis=1).columns)
```

```python
missing_info= list(data.columns[data.isnull().any()]) 
missing_info
```

```python
for col in missing_info:
    num_missing = data[data[col].isnull() == True].shape[0]
    print('number missing for column {}: {}'.format(col, num_missing))
```

```python
for col in missing_info: 
    percent_missing = data[data[col].isnull() == True].shape[0] / data.shape[0]
    print('percent missing for column {}: {}'.format( col,percent_missing))
```
